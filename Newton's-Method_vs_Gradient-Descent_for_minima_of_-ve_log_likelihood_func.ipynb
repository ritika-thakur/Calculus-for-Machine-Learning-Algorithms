```python
import numpy as np

import scipy.stats as s
```

```python
sample = np.random.rayleigh(scale=48, size=(40,))
```

```python
sigma_be = np.sqrt(np.mean(sample**2)/2)
```

```python
sigma_be
```

```python
sigma_pool = np.arange(start=30.78, stop=100.78, step = 0.5)
```

```python
sigma_pool
```

```python
def log_like_func(sigma_initial):
    
    return -(np.sum(s.rayleigh.logpdf(sample, scale=sigma_initial)))
```

```python
log_like_func_values = []

for sigma in sigma_pool:
    
    log_like_func_values.append(log_like_func(sigma))
```

```python
log_like_func_values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))

plt.plot(sigma_pool, log_like_func_values)

plt.xlabel("Different point estimates of sigma")

plt.ylabel("Different values of log likelihood function")
```

Minima of this negative log likelihood function of rayleigh distribution curve is 48.825

```python
def f_prime(sigma_initial):
    
    return np.sum((2*sigma_initial**2-sample**2)/sigma_initial**3)
```

```python
def f_double_prime(sigma_initial):
    
    return np.sum((3*sample**2 - 2*sigma_initial*2)/sigma_initial**4)
```

# NEWTON'S METHOD

```python
sigma_initial=35

epsilon = 10 ** (-5)
```

```python
while(abs(f_prime(sigma_initial)) > epsilon):
    
    sigma_final  = sigma_initial - (f_prime(sigma_initial)/f_double_prime(sigma_initial))
    
    print("The current guess of minima is {} and the value of negative log likelihood function is {}".format(sigma_initial, log_like_func(sigma_initial)))
    
    sigma_initial = sigma_final
    
print("\n\nConvergence is achieved, minima is {} and the minimum value of negative log likelihood funciton is {}".format(sigma_initial, log_like_func(sigma_initial)) )
```

# GRADIENT DESCENT

```python
sigma_initial = 35

mu = 1

iteration = 0
```

```python
while(abs(f_prime(sigma_initial)) > epsilon):
    
    sigma_final = sigma_initial - (mu * f_prime(sigma_initial))
    
    print("Iteration # {}, minima is {} and negative log likelihood is{}".format(iteration, sigma_initial, log_like_func(sigma_initial)))
    
    sigma_initial = sigma_final
    
    iteration = iteration + 1

print("\n\n Iteration # {}, minima is {} and  negative log likelihood is {}".format(iteration, sigma_initial, log_like_func(sigma_initial)) )
```
