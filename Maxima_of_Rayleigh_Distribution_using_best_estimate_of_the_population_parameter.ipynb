```python
import numpy as np

import scipy.stats as s
```

```python
sample = np.random.rayleigh(scale=48, size=(40,))
```

```python
sigma_be = np.sqrt(np.mean(sample**2)/2)
```

```python
sigma_be
```

```python
sigma_pool = np.arange(start=30.0085, stop=100.0085, step = 0.5)
```

```python
sigma_pool
```

```python
def log_like_func(sigma_initial):
    
    return np.sum(s.rayleigh.logpdf(sample, scale=sigma_initial))
```

```python
log_like_func_values = []

for sigma in sigma_pool:
    
    log_like_func_values.append(log_like_func(sigma))
```

```python
log_like_func_values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))

plt.plot(sigma_pool, log_like_func_values)

plt.xlabel("Different point estimates of sigma")

plt.ylabel("Different values of log likelihood function")
```

Maxima of this rayleigh distribution curve is 48.9
