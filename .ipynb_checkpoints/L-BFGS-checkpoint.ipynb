```python
import numpy as np
```

```python
X_i = np.array([1,-1]).reshape(-1,1)
```

```python
X_i
```

```python
epsilon = 10**(-5)
```

```python
mu_i = 10**(-3)
```

```python
H_i = np.eye(2,2)
```

```python
I = np.eye(2,2)
```

```python
def f(X_i):
    
    return X_i[0,0]**2 + 5*(10)**(-10)*X_i[1,0]**2
```

```python
f(X_i)
```

```python
def del_f(X_i):
    
    return np.array([2*X_i[0,0], 10**(-9)*X_i[1,0]]).reshape(-1,1)
```

```python
iterations = 0

while(np.linalg.norm(del_f(X_i)) > epsilon):
    
    p_i = -np.matmul(H_i,del_f(X_i))
    
    X_f = X_i + (mu_i*p_i)
    
    if f(X_f) > f(X_i):
        
        mu_i = np.random.uniform(low = 10**(-4), high=10**(-1))
        
        X_f = X_i + (mu_i * p_i)
        
    s_i = X_f - X_i
    
    y_i = del_f(X_f) - del_f(X_i)
    
    H_i  = np.matmul(np.matmul((I - np.matmul(s_i,y_i.T)/np.matmul(y_i.T,s_i)),H_i),(I-(np.matmul(y_i,s_i.T)/np.matmul(y_i,s_i.T)))) + (np.matmul(s_i,y_i.T)/np.matmul(y_i.T,s_i))
    
    X_i = X_f
    
    
    print("Iteration # {}, Current guess {}, Min Function Value {}".format(iterations, X_i[:,0],f(X_i)))
    
    iterations = iterations+1

print("Convergence achieved, Iteration # {}, Minima is {}, Minimum value {}".format(iterations, X_i[:,0],f(X_i)))
```
